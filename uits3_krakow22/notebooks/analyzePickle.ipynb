```python
%jsroot on

nBins = {"x-Axis" : 300, "y-Axis" : 150}
nRange = {"x-Axis" : 15, "y-Axis" : 7.5}
axis = {"x-Axis":0, "y-Axis":1}

detector1 = "ALPIDE_0"
detector2 = "ALPIDE_0"
axis1="x-Axis"
axis2="y-Axis"
lowerClustSizThresh = 7  #thresholds for how big to make the file for analysis
upperClustSizThresh = 15
clustNumThresh = 1
    
hCorr = ROOT.TH2F("Corr_"+detector1+"_"+axis1+"_"+detector2+"_"+axis2,"Correlation", nBins.get(axis1),-nRange.get(axis1),nRange.get(axis1),nBins.get(axis2),-nRange.get(axis2),nRange.get(axis2))
hCorr.SetXTitle(detector1 + " - " + axis1)
hCorr.SetYTitle(detector2 + " - " + axis2)

for event in readEvents(path, nEvents=-1):
    clustersDetector1 = event.selectDetector(detector1)
    clustersDetector2 = event.selectDetector(detector2)
    selectedClusters1 = [cluster for cluster in clustersDetector1 if upperClustSizThresh > cluster.size > lowerClustSizThresh]
    selectedClusters2 = [cluster for cluster in clustersDetector2 if upperClustSizThresh > cluster.size > lowerClustSizThresh]
    if len(selectedClusters1) <= clustNumThresh and len(selectedClusters2) <= clustNumThresh:
        for cluster1 in selectedClusters1:
            for cluster2 in selectedClusters2:
                hCorr.Fill(cluster1.localPos[axis.get(axis1)],cluster2.localPos[axis.get(axis2)])
                

canvas = ROOT.TCanvas()
hCorr.Draw("COLZ")
canvas.Draw()
```

```python
%jsroot on

nBins = {"x-Axis" : 300, "y-Axis" : 150}
nRange = {"x-Axis" : 30, "y-Axis" : 15}
axis = {"x-Axis":0, "y-Axis":1}


detector1 = "ALPIDE_4"#anchor detector
detector2 = "ALPIDE_3"

axis1="x-Axis"
axis2="y-Axis"

#[0.1723-0.02306, -0.9259+0.1166, 0] between ALPIDEs 3 & 4

lowerClustSizThresh = 7  #thresholds for how big to make the file for analysis
upperClustSizThresh = 15
clustNumThresh = 1

#calibrations made for 1e6 events on data file ending in 928

displacements0 = {
    "ALPIDE_0" : [0,0,0], # anchor
    "ALPIDE_1" : [1.156+0.05531,0.2837-0.03655,0],
    "ALPIDE_2" : [0.06475+0.2488,0.08900-0.03316,0],
    "ALPIDE_3" : [1.426-0.1012,0.7004-1.116,0],
    "ALPIDE_4" : [0.5261+0.5159,1.102-1.372,0],
}

displacements1 = {
    "ALPIDE_0" : [-1.211,-0.2472,0], 
    "ALPIDE_1" : [0,0,0],# anchor
    "ALPIDE_2" : [-0.7214,0.4154,0],
    "ALPIDE_3" : [0.4301,-0.02330,0],
    "ALPIDE_4" : [0.48375,0.01058,0],
}

displacements2 = {
    "ALPIDE_0" : [-0.3135,-0.05585,0], 
    "ALPIDE_1" : [0.7214,-0.4154,0],
    "ALPIDE_2" : [0,0,0],# anchor
    "ALPIDE_3" : [0.8794,-0.1445,0],
    "ALPIDE_4" : [0.9993,-0.1880,0],
}

displacements3 = {
    "ALPIDE_0" : [-1.321,0.4156,0], 
    "ALPIDE_1" : [-0.4301,0.02330,0],
    "ALPIDE_2" : [-0.8794,0.1445,0],
    "ALPIDE_3" : [0,0,0],# anchor
    "ALPIDE_4" : [-0.1492,0.8093,0],
}

displacements4 = {
    "ALPIDE_0" : [-1.042,0.2698,0], 
    "ALPIDE_1" : [-0.4837,-0.01058,0],
    "ALPIDE_2" : [-0.9960,0.1885,0],
    "ALPIDE_3" : [0.1492,-0.80930,0],
    "ALPIDE_4" : [0,0,0],# anchor
}

hCorr = ROOT.TH2F("Corr_"+detector1+"_"+axis1+"_"+detector2+"_"+axis2,"Correlation of "+detector1+" with "+detector2, nBins.get(axis1),-nRange.get(axis1),nRange.get(axis1),nBins.get(axis2),-nRange.get(axis2),nRange.get(axis2))
hCorr.SetXTitle(axis1 + " / mm")
hCorr.SetYTitle(axis2 + " / mm")

nEvents=1e6

#progress bar
f = IntProgress(min=0, max=nEvents, description="Processing:")
display(f)
for index, event in enumerate(readEvents(path, nEvents)):
    clustersDetector1 = event.selectDetector(detector1)
    clustersDetector2 = event.selectDetector(detector2)
    
    selectedClusters1 = [cluster for cluster in clustersDetector1 if upperClustSizThresh > cluster.size > lowerClustSizThresh]
    selectedClusters2 = [cluster for cluster in clustersDetector2 if upperClustSizThresh > cluster.size > lowerClustSizThresh]
    
    if len(selectedClusters1) <= clustNumThresh and len(selectedClusters2) <= clustNumThresh:
        for cluster1 in selectedClusters1:
            cluster1.alignCluster(displacements4.get(detector1))
            for cluster2 in selectedClusters2:
                cluster2.alignCluster(displacements4.get(detector2))
                
                hCorr.Fill(cluster1.localPos[0]-cluster2.localPos[0],cluster1.localPos[1]-cluster2.localPos[1])

    
    if index%100 == 0 :
        f.value += 100
        sys.stdout.write('\r')
        sys.stdout.write("[ %s ] events processed (%.2f%% complete)" % (index, 100*index/nEvents))
        sys.stdout.flush()
        #sleep(0.25)        
f.bar_style = "success"
canvas = ROOT.TCanvas()
hCorr.Draw("COLZ")
canvas.Draw()
```
